```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
trainset = pd.read_csv('Portugese Bank Data - TRAIN(1).csv')
print(trainset.shape)
trainset.head()
```

```python
all_dummies = pd.get_dummies(trainset, columns=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y' ], drop_first=True)
```

```python
trainset = pd.concat([trainset, all_dummies], axis=1)
```

```python
trainset.head()
```

```python
trainset.to_csv("trainset.csv",index=False)
```
