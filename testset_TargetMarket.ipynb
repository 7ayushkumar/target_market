```python
import pandas as pd
```

```python
import numpy as np
```

```python
testset = pd.read_csv('Portugese Bank Data - TEST(1).csv')
print(testset.shape)
testset.head()
```

```python
all_dummies = pd.get_dummies(testset, columns=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y' ], drop_first=True)
```

```python
testset = pd.concat([testset, all_dummies], axis=0)
```

```python
testset.head()
```

```python
testset.to_csv("testset.csv",index=False)
```
